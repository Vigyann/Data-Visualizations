```python
import matplotlib.pyplot as plt

def get_grade():
    while True:
        try:
            grade = int(input("Enter the grade (0-100): "))
            if 0 <= grade <= 100:
                return grade
            else:
                print("Grade must be between 0 and 100. Try again.")
        except ValueError:
            print("Invalid input. Please enter a numerical value.")

def fun():
    subjects = []
    grades = []

    while True:
        subject = input("Enter a subject name (or '4' to stop): ")
        if subject == '4':
            break
        grade = get_grade()

        subjects.append(subject)
        grades.append(grade)

   
    plt.figure(figsize=(10, 6))
    plt.bar(subjects, grades, color='skyblue')
    plt.xlabel('Subject')
    plt.ylabel('Grade')
    plt.title('Student Grades by Subject')
    plt.xticks(rotation=45)
    plt.tight_layout()  
    plt.show()


fun()
```

```python
def collect_data():
    data={'Age':[],'Taste_Rating':[]}

    while True:
        age=input("Enter Age Group( or type 'done to finish)")
        if age.lower()=='done':
            break
        taste_rating=input("Enter Taste Rating (1-5) :")

        data['Age'].append(age)
        data['Taste_Rating'].append(float(taste_rating))

    return pd.DataFrame(data)

df=collect_data()

plt.figure(figsize=(10,6))
sns.swarmplot(x='Age',y='Taste_Rating',data=df)
plt.title('Taste Ratings by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Taste Ratings')
plt.show()




```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def collect_data():
    data = {'Fitness_Level': [], 'Activity_Rating': []}

    while True:
        fitness_level = input("Enter Fitness Level (Beginner, Intermediate, Advanced) or type 'done' to finish: ").lower()
        
        if fitness_level == 'done':
            break
        
        if fitness_level not in ['beginner', 'intermediate', 'advanced']:
            print("Invalid Fitness Level.")
            continue
        
        try:
            activity_rating = int(input("Enter Activity Rating (1-5): "))
            if not (1 <= activity_rating <= 5):
                print("Invalid rating.")
                continue
        except ValueError:
            print("Invalid input. Please enter a valid number.")
            continue
        
        data['Fitness_Level'].append(fitness_level.capitalize())
        data['Activity_Rating'].append(float(activity_rating))

    return pd.DataFrame(data)

df = collect_data()

df['Fitness_Level'] = pd.Categorical(df['Fitness_Level'], categories=['Beginner', 'Intermediate', 'Advanced'])

plt.figure(figsize=(10, 6))
sns.swarmplot(x='Fitness_Level', y='Activity_Rating', data=df)
plt.title('Activity Ratings by Fitness Level')
plt.xlabel('Fitness Level')
plt.ylabel('Activity Rating')
plt.show()
print("No data collected to visualize.")
```
