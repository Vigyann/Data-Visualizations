```python
import matplotlib.pyplot as plt

new_activities=int(input("Enter the number of activities :"))

activities=[]
hours=[]

for i in range(new_activities):
    activity=input(f"Enter Activity{i+1}name: ")
    time_spent=float(input(f"Enter hours spent on {activity}:"))
    activities.append(activity)
    hours.append(time_spent)

plt.pie(hours,labels=activities,autopct='%1.1f%%',startangle=140)

plt.title("Time allocation in a day")
plt.axis('equal')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

app_data=[
    {'category':'Productivity','rating':4.1},
     {'category':'Productivity','rating':4.7},
    {'category':'Productivity','rating':3.9},
     {'category':'Productivity','rating':4.3},
      {'category':'Entertainment','rating':4.8},
       {'category':'Entertainment','rating':4.5},
    {'category':'Entertainment','rating':3.7},
    {'category':'Entertainment','rating':4.2},
    {'category':'Health','rating':4.9},
    {'category':'Health','rating':4.3},
    {'category':'Health','rating':4.0},
    {'category':'Health','rating':3.8},
]

#converting data to lists
categories=[app['category'] for app in app_data]
ratings=[app['rating']for app in app_data]   #keyword arguments

plt.figure(figsize=(8,6))
sns.swarmplot(x=categories,y=ratings,palette="Set2")

plt.title('User Ratings of Mobile apps by cateogery')
plt.xlabel('App Category')
plt.ylabel(' User Ratings')
plt.xticks(rotation=45)
plt.grid(True,linestyle='--',alpha=0.6)
plt.show()
```

```python
num_subjects=int(input("Enter no of subjects :"))
marks=[]
subject_names=[]

for i in range(num_subjects):
    subject_name=input(f"Emter the name of the subject {i+1}:")
    mark=float(input(f"Enter the marks abtaine din {subject_name}"))
    subject_names.append(subject_name)
    marks.append(mark)

total_marks=sum(marks)
max_marks=num_subjects*100
percentage=(total_marks/max_marks)*100

print(f"\nTotal Marks :{total_marks} out of {max_marks}")
print(f"Percentage :{percentage:.2f}%")


plt.figure(figsize=(10,6))
sns.barplot(x=subject_names,y=marks,palette='viridis')
plt.title('Marks obtained in each subject')
plt.xlabel('Subject')
plt.ylabel('Marks Obtained')
plt.ylim(0,100)# assuming each sub in range 1-100
plt.axhline(y=percentage,color='r',linestyle='--',label=f'Percentage :{percentage :2f}%')
plt.legend()
plt.show()
```

```python
celcius_temps=[]
farenheit_temps=[]

while True:
    temp=input("Enter temperature in celcius (or 'exit' to stop)")
    if temp.lower()=='exit':
        break
    celsius=float(temp)
    farenheit=(celsius*9/5)+32
    celcius_temps.append(celsius)
    farenheit_temps.append(farenheit)
    print(f"Temperature in farenheight : {farenheit:.2f}degree F")

plt.figure(figsize=(10,6))
sns.lineplot(x=celcius_temps,y=farenheit_temps,marker='o',color='blue',label='Fahrenheit')
plt.title('Celcius to fahrenheit conversion')
plt.xlabel('Celcius(DEgree c)')
plt.ylabel('Fahrenheit(Degree F)')
plt.grid()
plt.show()
```

```python
import time 
seconds=int(input("Enter time in seconds"))

remaining_time=[]

for i in range(seconds,0,-1):
    print(i)
    remaining_time.append(i)
    time.sleep(1)

print("Time's up!")

plt.figure(figsize=(10,6))
sns.barplot(x=list(range(1,seconds+1)),y=remaining_time,color='blue')
plt.title('Countdown Timer')
plt.xlabel('Seconds')
plt.ylabel('Remaining Time')
plt.xticks(range(1,seconds+1))
plt.ylim(0,seconds)
plt.grid(axis='y')
plt.show()
```
