```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data={
    'Aspect':['Design','Usability','Performance','Support','Price'],
    'Positive':[80,30,70,60,50],
    'Negative':[10,50,15,25,40]
}
df=pd.DataFrame(data)

plt.figure(figsize=(8,6))
sns.heatmap(df.set_index('Aspect'),annot=True,cmap='coolwarm')
plt.title('Customer Feedback sentiment analysis')
plt.show()
```

```python
data={
    'Department':['HR','Finance','IT','Marketing','Sales'],
    'Satisfaction':[4.2,3.8,4.5,4.0,3.5]
}
df=pd.DataFrame(data)

plt.figure(figsize=(8,6))
sns.barplot(x='Department',y='Satisfaction',data=df)
plt.title('Employeeb Satisfaction by department')
plt.show()
```

```python
data={
    'Destination':['Paris','Tokyo','New York','London','Sydney'],
    'Bookings':[150,200,300,250,100],
    'Trip duration (days)':[5,7,5,6,10]
}
df=pd.DataFrame(data)

df['Bookings']=df['Bookings'].astype(int)

df['Trip duration (days)']=df['Trip duration (days)'].astype(int)

plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='Bookings',y='Destination',size='Trip duration (days)',sizes=(100,500),legend=None)
plt.title('Destination Popularity Based on Bookings')
plt.xlabel('Number of Bookings')
plt.ylabel('Destination')
plt.show()
```

```python
data={
    'Category':['Electronics','Clothing','HomeGoods','Beauty','Sports'],
    'Total Sales':[30000,15000,20000,12000,18000],
    'Returns':[1500,800,1200,600,900],
    'Average Price':[300,50,100,40,90]
}
df=pd.DataFrame(data)
plt.figure(figsize=(12,10))

plt.subplot(3,1,1)
sns.barplot(data=df,x='Category',y='Total Sales',palette='viridis')
plt.title('Total sales by cateogery')
plt.ylabel("Total Sales ($)")

plt.subplot(3,1,2)
sns.lineplot(data=df,x='Category',y='Returns',color='orange',marker='o')
plt.title('Returns by cateogery')
plt.ylabel("Returns ($)")

plt.subplot(3,1,3)
sns.boxplot(data=df[['Average Price']])
plt.title('Average Price by cateogery')
plt.ylabel("Average Price ($)")

plt.tight_layout()
plt.show()
```

```python
data={
    'Source':['search','Social Media','Direct','Referral','Email'],
    'Visits':[5000,3000,2000,1500,1000],
    'Bounce Rate':[30,50,20,40,25],
    'Conversion Rate':[5,3,10,4,2]
}

df=pd.DataFrame(data)

plt.figure(figsize=(12,8))
plt.subplot(3,1,1)
sns.barplot(data=df,x='Source',y='Visits',palette='Set2')
plt.title('Website visits by source')
plt.ylabel('Number of visits')


plt.subplot(3,1,2)
sns.lineplot(data=df,x='Source',y='Bounce Rate',marker='o',color='red')
plt.title('Bounce Rate by source')
plt.ylabel('Bounce Rate(%)')


plt.subplot(3,1,3)
sns.violinplot(data=df[['Conversion Rate']])
plt.title('Distribution of conversion rates')
plt.ylabel('Conversion Rate(%)')

plt.tight_layout()
plt.show()
```

```python

data = {
    'Post_types': ['Image', 'Video', 'Texts', 'Audio', 'Image', 'Video', 'Texts'],
    'engagements': [120, 200, 300, 400, 200, 130, 40]
}

df = pd.DataFrame(data)

grouped_df = df.groupby('Post_types').mean().reset_index()

# Plotting
sns.barplot(data=grouped_df, x='Post_types', y='engagements')
plt.title('Total Engagements by Post Type')
plt.show()
```

```python
data={
    'Genre':['Action','Comedy','Horror','Scifi','Horror','Comedy','Comedy','Action','Scifi'],
    'Ratings':[8.5,7.2,6.7,9.6,7.2,4.3,5.6,9.0,6.0]
}
df=pd.DataFrame(data)
sns.boxplot(x='Genre',y='Ratings',data=df)
plt.title('Box plot ratings by genre')
plt.show()
```

```python
data={
    'durations':[30,45,60,20,50,70],
    'calories':[300,450,600,120,520,690]
}
df=pd.DataFrame(data)

sns.regplot(x='durations',y='calories',data=df)
plt.title('Regression line for calories vs duration')
plt.show()
```

```python

data = {
    'names': ['Alice', 'Bob', 'Sonu', 'Rahul', 'Vigyan', 'Om'],
    'Scores': [90, 85, 62, 56, 78, 96],
    'departments': ['HR', 'Sales', 'HR', 'Finance', 'IT', 'HR']
}
df = pd.DataFrame(data)
grouped_df = df.groupby('departments')['Scores'].mean().reset_index()
sns.barplot(x='departments', y='Scores', data=grouped_df)
plt.title('Average Scores by Department')
plt.show()
```
