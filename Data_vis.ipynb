```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('d5.csv')

data['Month'] = pd.Categorical(data['Month'], categories=['January','February', 'March', 'April', 'May', 'June','July', 'August','September', 'October', 'November', 'December'],
ordered=True)
data['Year'] = data['Year'].astype(str)

sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Month', y='Total_Sales', hue='Year', marker='o')
plt.title("Monthly Sales Trends Over Three Years")
plt.ylabel("Total Sales")
plt.xlabel("Month")
plt.legend(title="Year")
plt.show()
```

```python
data['Cumulative_Sales'] = data.groupby('Year')['Total_Sales'].cumsum()
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Month', y='Cumulative_Sales', hue='Year',marker='o', linewidth=2.5)
plt.fill_between(data['Month'], data['Cumulative_Sales'], alpha=0.2)
plt.title("Cumulative Sales Over Three Years")
plt.ylabel("Cumulative Sales")
plt.xlabel("Month")
plt.legend(title="Year")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Month', y='Total_Sales')
plt.title("Monthly Sales Distribution")
plt.ylabel("Total Sales")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.show()
```

```python
heatmap_data = data.pivot(index='Month', columns='Year', values='Total_Sales')
plt.figure(figsize=(8, 10))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu",cbar_kws={'label': 'Total Sales'})
plt.title("Sales Performance Heatmap Across Months and Years")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "d24.csv"
data = pd.read_csv(url)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='DiscountPercentage', y='SalesBeforeDiscount',hue='ProductCategory', style='ProductCategory', palette='Set1')
plt.title("Sales vs. Discount Percentage")
plt.xlabel("Discount Percentage")
plt.ylabel("Sales Before Discount")
plt.legend(title='Product Category', loc='best')
plt.show()

avg_sales = data[['SalesBeforeDiscount', 'SalesAfterDiscount']].mean()
avg_sales.plot(kind='bar', figsize=(8, 5))
plt.title("Average Sales Before and After Discount")
plt.ylabel("Average Sales")
plt.xticks([0, 1], ['Before Discount', 'After Discount'], rotation=0)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('d25.csv')

print(data.head())

plt.figure(figsize=(10,6))
sns.boxplot(x='AgeGroup', y='SatisfactionRating', data=data, hue='Gender')
plt.title('Box Plot: Satisfaction Ratings by Age Group and Gender')
plt.show()

plt.figure(figsize=(10,6))
sns.violinplot(x='ServiceSpeed', y='SatisfactionRating', data=data)
plt.title('Violin Plot: Satisfaction Ratings vs Service Speed')
plt.show()

correlation_matrix = data[['ServiceSpeed', 'ProductQuality', 'CustomerService','SatisfactionRating']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap: Correlation Between Satisfaction Factors')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('d26.csv')

print(data.head())

sns.pairplot(data[['AdvertisingSpend', 'ProductPrice', 'CustomerTraffic','Promotions', 'Sales']])
plt.suptitle('Pair Plot: Relationships Between Sales-Related Variables', y=1.02)
plt.show()

correlation_matrix = data[['AdvertisingSpend', 'ProductPrice','CustomerTraffic', 'Promotions', 'Sales']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f',linewidths=0.5)
plt.title('Heatmap: Correlation Matrix of Sales-Related Factors')
plt.show()

plt.figure(figsize=(8,6))
sns.regplot(x='AdvertisingSpend', y='Sales', data=data, line_kws={'color':'red'})
plt.title('Scatter Plot: Sales vs Advertising Spend')
plt.show()

plt.figure(figsize=(10,6))
sns.lineplot(x='Month', y='Sales', data=data, label='Sales')
sns.lineplot(x='Month', y='ProductPrice', data=data, label='Product Price',linestyle='--')
plt.title('Line Plot: Trends of Sales and Product Price Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import pandas pd
import seaborn as sns
import matplotlib.pyplot as plt

url="d8.csv"
data=pd.read_csv(url)

segment_sales=data.groupby('Customer_Segment')['Total_Sales'].sum()
segment_sales.plot(kind='pie',autopct='%1.1f%%',startangle=140,figsize=(8,8))
plt.title('percentsge distribution of total sales by customer segment')
plt.ylabel('')
plt.show()


#viz2
sns.barplot(x='Customer_Segment',y='Average_Spending',data=data,ci=None,palette='pastel')
plt.title('Average spending across customer segment')
plt.xlabel('customer segment')
plt.ylabel('Average spending')
plt.show()

#viz3
data['Age_Group_Mid']=data['Age_Group'].apply(lambda x: (int(x.split('-')[0])+int(x.split('-')[1]))//2 if '-' in x else 66)
sns.scatterplot(x='Age_Group_Mid',y='Average_Spending',hue='Customer_Segment',data=data,palette='deep')
plt.title('Relationship between customer age and average spending')
plt.xlabel('Age (midpoint)')
plt.ylabel('Average spending')
plt.legend(title='Customer Segment')
plt.show()


#viz4
category_data=data.groupby('Customer_Segment')[['Product_Category_A','Product_Category_B','Product_Category_C']].sum()
sns.heatmap(category_data,annot=True,fmt='d',cmap='YlGnBu')
plt.title('Purchasing Behaviour across segments and product categories')
plt.xlabel('Product Categories')
plt.ylabel('Customer  segment')
plt.show()
```
