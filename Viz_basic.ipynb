```python
import matplotlib.pyplot as plt

x=['A','B','C','D']
y=[10,15,7,12]

#create a bar plot
plt.bar(x,y,color='skyblue')

plt.title('Basic Bar plot')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

```python
sizes=[15,30,45,10]
labels=['Apples','Bananas','Cherries','Dates']
colors=['gold','yellowgreen','lightcoral','lightskyblue']

plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=140)

plt.title('Fruit Distribution')

#equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.show()
```

```python
months=['January','February','March','April']
milestones_achieved=[1,2,3,4]

plt.figure(figsize=(10,6))
plt.plot(months,milestones_achieved,marker='o',color='b',linestyle='-',linewidth=2,markersize=8)

plt.title('Project Progress over time')
plt.xlabel('Month')
plt.ylabel('NO of milestones completed')

#displaying the grid
plt.grid(True)

plt.show()
```

```python
#create subplots
fig,ax=plt.subplots(2,2,figsize=(10,8))

ax[0,0].plot([1,2,3],[4,5,6])
ax[0,0].set_title('Plot 1')

ax[0,1].scatter([1,2,3],[4,5,6])
ax[0,1].set_title('Plot 2')

ax[1,0].bar(['A','B','C'],[10,15,20])
ax[1,0].set_title('Plot 3')

ax[1,1].hist([1,2,2,3,4,4,5],bins=5)
ax[1,1].set_title('Plot 4')

plt.tight_layout()

plt.show()



```

```python
x=[1,2,3,4,5]
y1=[2,3,5,7,11]
y2=[1,2,4,6,8]

#create an area plot
plt.fill_between(x,y1,color='skyblue',alpha=0.7,label='Area1')
plt.fill_between(x,y2,color='salmon',alpha=0.5,label='Area2')

plt.title('Area plot example')
plt.xlabel('X-axis label')
plt.ylabel('Y- axis label')
plt.legend()

plt.show()

```

```python
x=[1,2,3,4,5]
y=[10,20,25,15,30]
labels=['A','B','c','D','E']

plt.scatter(x,y,color='blue')

#annoatate each  pont with a label
for i in range(len(x)):
    plt.annotate(labels[i],(x[i],y[i]),textcoords="offset points",xytext=(0,10),ha='center')

plt.title('Scatter plot with annotations')
plt.xlabel('X-axis label')
plt.ylabel('Y- axis label')

plt.show()

```

```python
x=[1,2,3,4,5]
y=[10,20,25,15,30]
labels=['A','B','c','D','E']

plt.scatter(x,y,color='blue')

#annoatate each  pont with a label
for i in range(len(x)):
    plt.annotate(labels[i],
                 (x[i],y[i]),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center',
                 color='red',
                 arrowprops=dict(facecolor='black',shrink=0.02))

plt.title('Scatter plot with  advanced annotations')
plt.xlabel('X-axis label')
plt.ylabel('Y- axis label')

plt.show()

```

```python
x=[1,2,3,4,5]
y=[10,20,25,15,30]
labels=['A','B','c','D','E']

plt.scatter(x,y,color='blue')

#annotate the point to label c
point_index=2 #index of the point to annoatate

plt.annotate(labels[point_index],(x[point_index],y[point_index]),textcoords="offset points",xytext=(10,10),ha='center',fontsize=12,color='red',bbox=dict(facecolor='yellow',alpha=0.7),arrowprops=dict(facecolor='black',arrowstyle='->'))

plt.title('Scatter plot with single point annotations')
plt.xlabel('X-axis label')
plt.ylabel('Y- axis label')

plt.show()

```
